```python
import matplotlib.pyplot as plt
import numpy as np
```

 **side by side bar graph**
 

```python
x=(1,2,3,4,5,6)
y=(1110,1120,1310,1140,1150,1160)
z=(112,214,316,418,160,712)
width=0.2
p=np.arange((len(x)))

plt.bar(p,y,width,color="r",label="1")
plt.bar(p+width,z,width,color="y",label="2")
plt.xticks(p+0.1,x)
plt.legend()
plt.show()
```

**pie chart**


```python
x=(10,20,30,40)
name=("aam","jaam","peeple","kaashi")
ex=(0.3,0.1,0,0)
plt.pie(x,labels=name,explode=ex,autopct="%0.1f%%")

plt.legend()
```

```python
x=(1,2,3,4,5,6)
y=(1110,1120,1310,1140,1150,1160)
z=(112,214,316,418,160,712)
plt.style.use('seaborn')
plt.subplot(2,2,1)
width=0.2
p=np.arange((len(x)))

plt.bar(p,y,width,color="r",label="1")
plt.bar(p+width,z,width,color="y",label="2")
plt.xticks(p+0.1,x)
plt.legend()


x=(10,20,30,40)
plt.subplot(2,2,4)
name=("aam","jaam","peeple","kaashi")
ex=(0.3,0.1,0,0)
plt.pie(x,labels=name,explode=ex,autopct="%0.1f%%")

plt.legend()
```
